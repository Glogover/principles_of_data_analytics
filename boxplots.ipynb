**BOXPLOTS**

```python
# Plotting.
import matplotlib.pyplot as plt

# Numerical arrays.
import numpy as np
```

**Continuous vs Discrete**

```python
plt.bar # bar plot
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

plt.bar(fruits, counts)
```

**CREATING DATA**

```python
# Random number generator.
rng = np.random.default_rng(12345)

# Show
rng
```

**GENERATING SOME RANDOM NORMAL DATA**

```python
# GROUP 1: Mean = 75, Std Dev = 10, Size = 30
group_1 = rng.normal(75, 10, 30)

# Show
print(group_1)
```

**Bar Charts vs Histograms**

```python
# List of unique values and their counts
unique, counts = np.unique(group_1, return_counts=True)

# Show
unique, counts
```

```python
plt.bar(unique, counts)
```

```python
# Histogram
plt.hist(group_1, bins = 30, edgecolor = 'black')
```

**MORE GROUPS**

```python
# GROUP 2: Mean = 65, Std Dev = 12, Size = 30
group_2 = rng.normal(65, 12, 30)

# Show
print(group_2)
```

```python
# GROUP 3: Mean = 85, Std Dev = 8, Size = 30
group_3 = rng.normal(85, 8, 30)

# Show
print(group_3)
```

```python
# Histograms
plt.hist(group_1, edgecolor = 'black', label ='Group 1', alpha = 0.5) # alpha = transparency
plt.hist(group_2, edgecolor = 'black', label ='Group 2', alpha = 0.5)
plt.hist(group_3, edgecolor = 'black', label ='Group 3', alpha = 0.5)
plt.legend()
```

```python
# Create a numpy array
data = np.array([group_1, group_2, group_3])

# Show
print(data)
```

```python
plt.hist(data.T)
```

```python
# Mean
np.mean(group_1)
```

```python
# Means
np.mean(data, axis=1)
```

```python
# Means
print(f'Means:               {np.mean(data, axis=1)}')

# Standard deviations
print(f'Standard deviations: {np.std(data, axis=1)}')

# Medians
print(f'Medians:             {np.median(data, axis=1)}')

# Maxima
print(f'Maxima:              {np.max(data, axis=1)}')

# Minima
print(f'Minima:              {np.min(data, axis=1)}')

```

```python
# Q1 (the first quartile)
np.quantile(data, 0.25, axis=1)
```

```python
# Create figure, axis
fig, ax = plt.subplots()

# Create boxplot
ax.boxplot(group_1) # The boxploy shows that the data is skewed
```

```python
# Create figure, axis
fig, ax = plt.subplots()

# Create boxplot
ax.boxplot(data.T); # T = transpose
```

**TILES AND AXIS LABELS**

```python
# Create figure, axis.
fig, ax = plt.subplots()

# Create boxplot.
ax.boxplot(data.T);

# Title.
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html
ax.set_title("Side-by-Side Boxplots of Test Scores", fontsize=16)

# Axis labels.
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html
ax.set_xlabel("Groups", fontsize=12)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html
ax.set_ylabel("Scores", fontsize=12)

# Set names of x-axis ticks.
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticks.html
ax.set_xticks([1, 2, 3], ["Group 1", "Group 2", "Group 3"], fontsize=10)

# Add a grid.
ax.grid(axis="y", linestyle="--", alpha=0.7)
```
