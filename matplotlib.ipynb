**matplotlib**

```python
# Plots
import matplotlib.pyplot as plt
```

```python
# Numerical arrays
import numpy as np
```

**Plotting Functions**

```python
# X values
x = np.arange(0.0, 10.0, 0.1)

# Show
x
```

```python
# X values
x = np.linspace(0.0, 10.0, 101) # The third value is the number of points that you want. This time 10 is included.

# Show
x
```

```python
# Y values
y = x**2  # x squared (x raised to the power 2)

# Show
y
```

```python
# Plot x vs y
plt.plot(x, y)
```

```python
# Plot x vs y
plt.plot(x, y);
```

```python
# Plot x vs y
plt.plot(x, y, label='y = x^2')

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, '.', label='y = x^2')

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, '-', label='y = x^2')

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, 'r-', label='y = x^2') # Red

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, 'g-', label='y = x^2') # Green

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, 'b-', label='y = x^2') # Blue

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y
plt.plot(x, y, 'k-', label='y = x^2') # Black

plt.legend();

plt.xlabel('x')
plt.ylabel('x^2')

plt.title('A Lovely Plot')
```

```python
# Plot x vs y.
plt.plot(x, y, label='$y = x^2$') # LaTeX format (Lay-tech)

# Add a legend.
plt.legend();

# Add labels.
plt.xlabel('x')
plt.ylabel('x^2')

# Add a title
plt.title('A Lovely Plot')
```

**COMPARING PLOTS**

```python
# Domain of x values.
x = np.linspace(0.0, 10.0, 101)

# x squared.
y1 = x**2

# 2 to the power of x.
y2 = 2**x
```

```python
# Plot x vs y1.
plt.plot(x, y1, label = '$y = x^2$')

# Plot x vs y2.
plt.plot(x, y2, label = '$y = 2^x$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Two Lovely Plots');

```
