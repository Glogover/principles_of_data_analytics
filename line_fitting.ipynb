## Data

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)
```

```python
# Use polyfit to fit a line to the data.
np.polyfit(x, y, 1)
```

```python
# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)
```

```python
# Show

m, c
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color= 'red');
```

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x + np.random.normal(0.0, 0.3, size=x.size) # + "noise"
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)
```

```python
# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color= 'red');
```

## All Together

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x + np.random.normal(0.0, 1.0, size=x.size) # + "noise", mean = 0, std = 3.0

# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c

# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color= 'red')

ax.set_title(f'y = {m:.2f}x + {c:.2f}');
```
